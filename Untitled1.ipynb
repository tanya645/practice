```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'Name': ['Aman', 'Riya', 'Rahul', 'Neha', 'Arjun', 'Simran', 'Karan', 'Priya'],
    'Maths': [78, 92, 67, 85, 74, 90, 60, 88],
    'Science': [85, 88, 72, 80, 69, 91, 65, 84],
    'English': [82, 90, 70, 88, 75, 89, 63, 86]
}

df = pd.DataFrame(data)
df
```

```python
df.head()
df.describe()
```

```python
df[['Maths','Science','English']].mean().plot(kind='bar')
plt.title("Average Marks")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()
```
